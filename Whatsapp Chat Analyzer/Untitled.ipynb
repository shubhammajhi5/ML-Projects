```python
import numpy as np
import pandas as pd
import re
```

```python
f = open('group_chat.txt', 'r', encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
## year have four digits and am pm in small

pattern = '\d{1,2}/\d{1,2}/\d{1,4},\s\d{1,2}:\d{1,2}\s[AaPp][Mm]\s-\s'
```

```python
messages = re.split(pattern, data)[1:]

messages
```

```python
dates = re.findall(pattern, data)

dates
```

```python
print('messages', len(messages))
print('dates', len(dates))
```

```python
df = pd.DataFrame({'user_message' : messages, 'date' : dates})

df.head()
```

```python
# changed format code %y --> %Y 
# and position of date and month %m/%d/%y --> %d/%m/%Y

df['date'] = pd.to_datetime(df['date'], format = '%d/%m/%Y, %I:%M %p - ')

df.head()
```

```python
df['user_message'] = df['user_message'].str.strip()
df.head()
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('chat_notification')
        messages.append(entry[0])
        

df['users'] = users
df['messages'] = messages
df.drop(columns = ['user_message'], inplace = True)

df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['date'] = df['date'].dt.date

df.head()
```
