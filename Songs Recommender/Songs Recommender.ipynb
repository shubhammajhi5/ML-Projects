```python
import pandas as pd
import numpy as np
```

```python
songs = pd.read_csv('bolly_songs.csv', encoding = 'cp1252')
```

```python
songs.head()
```

```python
songs.shape
```

```python
songs.info()
```

```python
songs.isnull().sum()
```

```python
songs[songs['Singer/Artists'].isnull() | songs['Album/Movie'].isnull()]
```

```python
songs.dropna(inplace = True)
```

```python
songs.isnull().sum()
```

```python
songs.shape
```

```python
songs.columns
```

```python
songs.rename(columns = {'Song-Id' : 'id',
                        'Song-Name' : 'song',
                        'Singer/Artists' : 'artist',
                        'Genre' : 'genre',
                        'Album/Movie' : 'movie',
                        'User-Rating' : 'rating'}, inplace = True)
```

```python
songs.head()
```

```python
songs.duplicated().sum()
```

```python
songs['artist'] = songs['artist'].apply(lambda x : x.split(','))
songs['artist']
```

```python
songs['artist'] = songs['artist'].apply(lambda x : [i.replace(' ', '') for i in x])
songs['artist']
```

```python
songs['genre'] = songs['genre'].apply(lambda x : x.split())
songs['genre']
```

```python
songs['movie'] = songs['movie'].apply(lambda x : x.replace(' ', ''))
songs['movie']
```

```python
songs['movie'] = songs['movie'].apply(lambda x : x.split())
songs['movie']
```

```python
songs['rating'] = songs['rating'].astype('str').apply(lambda x : x.split())
songs['rating']
```

```python
songs['tags'] = songs['artist'] + songs['genre'] + songs['movie'] + songs['rating']
```

```python
songs.head()
```

```python
new_df = songs[['id', 'song', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : ' '.join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
new_df.head()
```

```python
new_df.reset_index(inplace = True)
new_df.drop(columns = ['index'], inplace = True)
new_df.head()
```

```python
new_df.tail()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x : x[1])[1:6]
```

```python
def recommend(song):
    song_index = new_df.loc[new_df['song'] == song].index[0]
    distances = similarity[song_index]
    songs_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x : x[1])[1:11]
    
    for i in songs_list:
        print(new_df.loc[i[0], 'song'])
```

```python
recommend('Lag Ja Gale')
```
