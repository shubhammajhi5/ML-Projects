## Popularity Based Recommender

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')

ratings_with_name.head()
```

```python
# we will recommend the books based on the ones having rating greater than the average rating. 
# also we will consider only those books which have more than 250 ratings (by people) in it.

num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.head()
```

```python
num_rating_df.rename(columns = {'Book-Rating' : 'num_ratings'}, inplace = True)
```

```python
num_rating_df.head()
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_ratings'}, inplace = True)
avg_rating_df.head()
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popular_df.head()
```

```python
popular_df = popular_df.loc[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending = False).head(50)
popular_df.head()
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'num_ratings', 'avg_ratings', 'Image-URL-M']]

popular_df.head()
```

```python
import pickle

pickle.dump(popular_df, open('popular.pkl', 'wb'))
```

## Collaborative Filtering based Recommender

```python
# we will utilise the books vs users matrix for this recommender
# we will only consider the users who have given reviews for over 200 books
# and we will consider only those books which recieved over 50 reviews
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >= 200
```

```python
valid_users = x[x].index
```

```python
filtered_rating = ratings_with_name.loc[ratings_with_name['User-ID'].isin(valid_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
valid_books = y[y].index
```

```python
final_rating = filtered_rating.loc[filtered_rating['Book-Title'].isin(valid_books)]
final_rating.head()
```

```python
pt = final_rating.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt.head()
```

```python
pt.fillna(0, inplace = True)
```

```python
pt.head()
```

```python
pt.shape
```

```python
# now we will consider each book in a 815 dimensional space

from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
similarity_scores[0]    ## this represents the similarity score of the first book with all the other 706 books
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Year of Wonders')
```

```python
pt.index
```

```python
def recommend_details(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x : x[1], reverse = True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommend_details('1984')
```
